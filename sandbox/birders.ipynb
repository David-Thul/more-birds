```python
cal = calendar.Calendar()
md = {}
for month in range(1,13):
    days = []
    for day in cal.itermonthdays(2017, month):
        if day !=0:
            days.append(day)
    md[month] = days
    
print(md)
```

```python
tx = []
for month in md.keys():
    
    for day in md[month]:
        print (str(month)+"-"+str(day))
        url = f"https://ebird.org/ws2.0/product/stats/US-TX/2017/{month}/{day}"       
        local_response = requests.get(url, headers=headers)
        data = local_response.text
        data = json.loads(data)
        data['month']=month
        data['day']=day
        tx.append(data)

tx
```

```python
import pandas as pd
tx_df = pd.DataFrame(tx)
tx_df['state']='TX'
tx_df
```

```python
tx_df.to_csv('birders/tx_17.csv', index=False)
```
