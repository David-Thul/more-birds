```python
import pandas as pd
import os
```

```python
dir_list = os.listdir('birders')
dir_list
```

```python
#make a list with all the CSV file names
files=[]
for i in range(0,len(dir_list)-1):
    if dir_list[i].endswith('.csv'):
        files.append(dir_list[i])
```

```python
df = pd.read_csv('birders/'+files[0])
```

```python
df['Year']=2016
df.head()
```

```python
for i in range (1,len(files)-1):
    add_df = pd.read_csv('birders/'+files[i])
    if files[i].endswith('16.csv'):
        add_df['Year']=2016
    else:
        add_df['Year']=2017
    df=pd.concat([df,add_df])
```

```python
#write consolidated CSV file
df.to_csv('birders/complete_data.csv')
```
